```python
import pandas as pd

qTM_df = pd.DataFrame(columns=["Structure", "Cluster", "Representative"])

with open("qTMcluster_results_queries.txt", "r") as qTM_file:
  qTM_lines = qTM_file.readlines()
  for cluster_nr in range(len(qTM_lines)):
    qTM_line = qTM_lines[cluster_nr][1:-5]
    print(qTM_line)
    qTM_cluster = qTM_line.split(".pdb\t/")
    for item in qTM_cluster:
      if item == qTM_cluster[0]:
        cluster_representative = item
      if ".pdb/" in item:
        item1 = item.split(".pdb/")[0]
        qTM_df.loc[len(qTM_df.index)] = [item1, cluster_nr, cluster_representative]
        item2 = item.split(".pdb/")[1]
        qTM_df.loc[len(qTM_df.index)] = [item2, cluster_nr, cluster_representative]
      else:
        qTM_df.loc[len(qTM_df.index)] = [item, cluster_nr, cluster_representative]
qTM_df.to_csv("qTMcluster_results.csv", index=False)
qTM_df
```
