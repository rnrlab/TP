```python
from Bio import Entrez, SeqIO
from Bio.Blast.Applications import NcbiblastpCommandline
import pandas as pd
import itertools
import os
import math

main = pd.read_csv("FINAL_results.csv")
main_dedup = main.copy()
genbankid = main_dedup["GenBankID"].to_list()
accession_numbers = []

for value in genbankid:
    if not (isinstance(value, float) and math.isnan(value)):
      accession_numbers.append(value)
    else:
      with open("UniProtKB.txt", "a") as UP:
        UP.write(f"{value}\n")

Entrez.email = "juancarlos.ramirezm@estudiante.uam.es"
def fetch_sequence(accession):
    if accession != "":
      print(f"Fetching {accession}")
      try:
          with Entrez.efetch(db="protein", id=accession, rettype="fasta", retmode="text") as handle:
              record = SeqIO.read(handle, "fasta")
              return record
      except Exception as e:
            print(f"Error fetching {accession}: {e}")
            print("No FASTA sequence available")
            print("+++")
            with open("Contig_masked.txt", "a") as contig:
              contig.write(f"{accession}\n")
            return None


# Fetch sequences and write them to FASTA files
sequences = {}
for acc in accession_numbers:
    seq_record = fetch_sequence(acc)
    sequences[acc] = seq_record
    try:
      with open(f"{acc}.fasta", "w") as output_handle:
          SeqIO.write(seq_record, output_handle, "fasta")
    except Exception as e:
      print(f"Error fetching {acc}: {e}")
      print("No FASTA sequence available")
      print("+++")
    print(f"{acc} saved to {output_handle}")
```

###UniProtKB ID

```python
import requests
import pandas as pd


def fetch_fasta_separate_files(uniprot_ids, output_dir="."):
    cnt = 0
    base_url = "https://www.uniprot.org/uniprot/"
    for uniprot_id in uniprot_ids:
        try:
            # Fetch the sequence in FASTA format
            response = requests.get(f"{base_url}{uniprot_id}.fasta")
            if response.status_code == 200:
                # Write each sequence to its own file
                output_file = f"{uniprot_id}.fasta"
                with open(output_file, "w") as fasta_file:
                    fasta_file.write(response.text)
                print(f"Successfully fetched and saved: {uniprot_id}")
            else:
                print(f"Failed to fetch {uniprot_id}: {response.status_code}")
                print("No FASTA sequence available")
                cnt = cnt + 1
                with open("BFVD_AFDB50_exclusive.txt", "a") as BFVD:
                  BFVD.write(f"{uniprot_id}\n")
        except Exception as e:
            print(f"Error fetching {uniprot_id}: {e}")
            print("No FASTA sequence available")
    print(cnt)

uniprot_ids = []
for i in range(len(main)):
    if isinstance(main["GenBankID"][i], float) and math.isnan(main["GenBankID"][i]):
        uniprot_ids.append(main["Hit"][i])
fetch_fasta_separate_files(uniprot_ids)
```

##BLASTp synonymous results examination

```python
import pandas as pd
import subprocess
import os

new_ids = []
id_dict = {}
id_replacement = pd.DataFrame(columns = ["UniProtID", "NEW_GenBankID", "%I", "E-value", "Bit-Score", "len(Qry)"])

command = "tar -xvf BLASTp_results.tar.gz"
subprocess.run(command, shell=True)

with open("ToBlast_ids.txt", "r") as b:
  b_list = b.read()
  b_list = b_list.split("\n")
  for uid in b_list:
    if uid != "" and os.path.isfile(f"{uid}.tsv"):
      with open(f"{uid}.tsv", "r") as tsv:
        for line in tsv:
          line = line.strip()
          if not line.startswith("#") and line != "":
            all_comments = False
            break
          else:
            all_comments = True
        if all_comments == False:
          print(f"Reading {uid}.tsv")
          blastp_result = pd.read_csv(f"{uid}.tsv", sep = "\t", comment = "#", header = None)
          new_ids.append(blastp_result[1][0])
          id_dict[uid] = blastp_result[1][0]
          print(f"Synonym: {blastp_result[1][0]}\n")
          with open("new_ids.txt", "a") as NI:
            NI.write(f'{blastp_result[1][0]}\n')
          id_replacement.loc[len(id_replacement.index)] = [uid, blastp_result[1][0], blastp_result[2][0], blastp_result[10][0], blastp_result[11][0], (int(blastp_result[9][0]) - int(blastp_result[8][0]) + 1)]
          id_replacement.to_csv("id_replacement.csv", index = False)
          id_replacement
        else:
          with open("Non-synonymous.txt", "a") as ns:
            print(f"Reading {uid}.tsv")
            print(f"NO BLASTp synonym was yielded for {uid}...\n")
            ns.write(f"{uid}\n")
    else:
      with open("Wrong.txt", "a") as w:
        w.write(f"{uid}\n")
```
