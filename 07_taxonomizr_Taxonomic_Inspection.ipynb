#**`taxonomizr`**

```r
install.packages("taxonomizr")
library(taxonomizr)
prepareDatabase('accessionTaxa.sql', getAccessions = FALSE)
```

```r
library(data.table)
library(dplyr)

main <- read.csv("FINAL_results.csv", header = TRUE, sep = ",")

#taxaId <- accessionToTaxa(main$GenBankID,"accessionTaxa.sql")
taxa <- getTaxonomy(unique(main$TaxID),'accessionTaxa.sql')
#print(taxa)
dataset <- as.data.frame(taxa)
dataset$TaxID <- as.numeric(row.names(dataset))
results_full <- merge(main, dataset, by="TaxID", all.x=TRUE)
results_full <- results_full %>% relocate(TaxID, .before = Database)
results_full
write.table(results_full, "FULL_results.csv", quote=FALSE,row.names = FALSE, sep=",")
```
