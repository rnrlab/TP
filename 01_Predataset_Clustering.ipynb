##PAIRWISE ALIGNMENT & CLUSTERING

```python
from Bio import Entrez, SeqIO
from Bio.Blast.Applications import NcbiblastpCommandline
import pandas as pd
import itertools
import os

accession_numbers = TP_list

# Set email for Entrez (required by NCBI)
Entrez.email = "juancarlos.ramirezm@estudiante.uam.es"

# Function to fetch protein sequences from GenBank
def fetch_sequence(accession):
    with Entrez.efetch(db="protein", id=accession, rettype="fasta", retmode="text") as handle:
        record = SeqIO.read(handle, "fasta")
        return record

# Fetch sequences and write them to FASTA files
sequences = {}
for acc in accession_numbers:
    seq_record = fetch_sequence(acc)
    sequences[acc] = seq_record
    with open(f"{acc}.fasta", "w") as output_handle:
        SeqIO.write(seq_record, output_handle, "fasta")

# Generate unique pairs without repetition
pairs = list(itertools.combinations(accession_numbers, 2))

# Create an empty DataFrame to store results
identity_matrix = pd.DataFrame(index=accession_numbers, columns=accession_numbers, dtype=float)

# Perform pairwise BLAST for each pair
for acc1, acc2 in pairs:
    # Run BLAST2seq
    blastp_cline = NcbiblastpCommandline(
        query=f"{acc1}.fasta", subject=f"{acc2}.fasta", outfmt=5, out="temp.xml"
    )
    stdout, stderr = blastp_cline()

    # Parse the BLAST output
    from Bio.Blast import NCBIXML
    with open("temp.xml") as result_handle:
        blast_record = NCBIXML.read(result_handle)
        if blast_record.alignments:  # Check if there is an alignment
            hsp = blast_record.alignments[0].hsps[0]
            percent_identity = (hsp.identities / hsp.align_length) * 100
        else:
            percent_identity = 0.0  # No alignment found

    # Fill the DataFrame
    identity_matrix.loc[acc1, acc2] = percent_identity
    identity_matrix.loc[acc2, acc1] = percent_identity

# Remove the temporary BLAST output file
os.remove("temp.xml")

# Print or save the DataFrame
print(identity_matrix)
identity_matrix.to_csv("pairwise_percent_identity.csv")
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

data = identity_matrix.fillna(100.00)

# Step 1: PCA Dimensionality Reduction
pca = PCA(n_components=2)
pca_result = pca.fit_transform(data)
pca_df = pd.DataFrame(pca_result, columns=['PC1', 'PC2'])

# Step 2: Clustering with KMeans
num_clusters = 12  # Fine-tune number of clusters
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
pca_df['Cluster'] = kmeans.fit_predict(pca_df)

rownames = [next((key for key, value in TP_Dataset.items() if value == row_name), row_name)
    for row_name in data.index]

pca_df['Label'] = rownames


# Step 3: Plotting the PCA result with clusters
plt.figure(figsize=(10, 8))
sns.scatterplot(
    x='PC1', y='PC2', hue='Cluster', data=pca_df, palette='viridis', s=100, alpha=0.8
)
for _, row in pca_df.iterrows():
    plt.text(row['PC1'], row['PC2'], row['Label'], fontsize=8, alpha=0.7)
plt.title('PCA with Clustering', fontsize=16)
plt.xlabel('Principal Component 1', fontsize=12)
plt.ylabel('Principal Component 2', fontsize=12)
#plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```
